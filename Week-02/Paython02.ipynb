```python
# important Libraries
import pandas as pd # Preprocessing, I/O, Cleaning
import numpy as np # Linear Algebra (Matrices Manipulation)
import matplotlib.pyplot as plt # Visualization
import seaborn as sns
sns.set_theme(style = "whitegrid")
```

```python
Emp = pd.read_csv('Employee.csv')
Edu = pd.read_csv('EducationLevel.csv')
Per = pd.read_csv('PerformanceRating.csv')
Rating = pd.read_csv('RatingLevel.csv')
Sat = pd.read_csv('SatisfiedLevel.csv')
```

```python
## Merge Education levels for readability
Emp = Emp.merge(Edu, left_on = 'Education', right_on = 'EducationLevelID', how = 'left')
```

```python
# Convert HireDate to datetime for time-series analysis
Emp['HireDate'] = pd.to_datetime(Emp['HireDate'])
Emp['HireYear'] = Emp['HireDate'].dt.year
```

### Q1: What is the gender distribution in the company?

```python
plt.figure(figsize = (12, 6))
Emp['Gender'].value_counts().plot.pie(autopct = '%1.1f%%', colors = sns.color_palette('pastel'))
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

### Q2: What is the age distribution of employees?

```python
plt.figure(figsize=(12, 6))
sns.histplot(data = Emp, x = 'Age', bins = 20, kde = True, color = 'skyblue')
plt.title('Age Distribution of Employees')
plt.show()
```

### Q3: How are employees distributed across different departments?

```python
plt.figure(figsize=(12, 6))
sns.countplot(data = Emp, x = 'Department', palette = 'viridis')
plt.title('Employee Count by Department')
plt.show()
```

### Q4: What is the overall attrition rate?

```python
plt.figure(figsize=(12, 6))
sns.countplot(data = Emp, x = 'Attrition', palette = 'Set2')
plt.title('Overall Attrition Distribution')
plt.show()
```

### Q5: How does attrition vary across different departments?

```python
plt.figure(figsize=(12, 6))
sns.countplot(data = Emp, x = 'Department', hue = 'Attrition', palette = 'Set1')
plt.title('Attrition by Department')
plt.show()
```

### Q6: What is the average salary by job role?

```python
plt.figure(figsize = (12, 6))
salary_role = Emp.groupby('JobRole')['Salary'].mean().sort_values(ascending = False)
sns.barplot(x = salary_role.values, y = salary_role.index, palette = 'mako')
plt.title('Average Salary by Job Role')
plt.xlabel('Average Salary ($)')
plt.show()
```

### Q7: Is there a gender pay gap? (Salary distribution by Gender)

```python
plt.figure(figsize = (12, 6))
sns.boxplot(data = Emp, x = 'Gender', y = 'Salary', palette = 'pastel')
plt.title('Salary Distribution by Gender')
plt.show()
```

### Q8: Does distance from home impact attrition?

```python
plt.figure(figsize = (12, 6))
sns.boxplot(data = Emp, x = 'Attrition', y = 'DistanceFromHome (KM)', palette = 'Set3')
plt.title('Impact of Distance from Home on Attrition')
plt.show()
```

### Q9: What is the distribution of education levels?

```python
plt.figure(figsize=(12, 6))
sns.countplot(data = Emp, y = 'EducationLevel', palette = 'cubehelix', order = Emp['EducationLevel'].value_counts().index)
plt.title('Distribution of Education Levels')
plt.show()
```

### Q10: How does business travel frequency vary?

```python
plt.figure(figsize = (12, 6))
sns.countplot(data = Emp, x = 'BusinessTravel', palette = 'magma')
plt.title('Business Travel Frequency')
plt.show()
```

### Q11: How does overtime affect attrition?

```python
plt.figure(figsize = (12, 6))
sns.countplot(data = Emp, x = 'OverTime', hue = 'Attrition', palette = 'Set1')
plt.title('Overtime vs. Attrition')
plt.show()
```

### Q12: Which states have the highest number of employees?

```python
plt.figure(figsize = (12, 6))
sns.countplot(data = Emp, x = 'State', palette = 'crest', order = Emp['State'].value_counts().index)
plt.title('Employee Count by State')
plt.show()
```

### Q13: What is the distribution of ethnicities?

```python
plt.figure(figsize = (12, 6))
sns.countplot(data = Emp, y = 'Ethnicity', palette = 'flare', order = Emp['Ethnicity'].value_counts().index)
plt.title('Ethnic Diversity')
plt.show()
```

### Q14: What is the historical hiring trend?

```python
plt.figure(figsize = (12, 6))
hiring_trend = Emp['HireYear'].value_counts().sort_index()
sns.lineplot(x = hiring_trend.index, y = hiring_trend.values, marker = 'o', color = 'purple')
plt.title('Hiring Trend Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Hires')
plt.show()
```

```python
merged_df = Emp.merge(Per, on = 'EmployeeID', how = 'inner')
```

### Q15: What is the average job satisfaction by department?

```python
plt.figure(figsize = (12, 6))
sns.barplot(data = merged_df, x = 'Department', y = 'JobSatisfaction', palette = 'rocket')
plt.title('Average Job Satisfaction by Department')
plt.ylim(1, 5) # Assuming satisfaction is out of 5
plt.show()
```

### Q16: How does work-life balance relate to attrition?

```python
plt.figure(figsize = (12, 6))
sns.boxplot(data = merged_df, x = 'Attrition', y = 'WorkLifeBalance', palette = 'Set2')
plt.title('Work-Life Balance vs Attrition')
plt.show()
```

### Q17: How do self-ratings compare to manager ratings?

```python
plt.figure(figsize = (12, 6))
sns.scatterplot(data = merged_df, x = 'SelfRating', y = 'ManagerRating', alpha = 0.5, color = 'darkgreen')
plt.title('Self Rating vs Manager Rating')
plt.show()
```

### Q18: What is the correlation between quantitative metrics?

```python
plt.figure(figsize = (12, 6))
numeric_cols = ['Age', 'Salary', 'DistanceFromHome (KM)', 'YearsAtCompany', 'JobSatisfaction', 'WorkLifeBalance']
corr_matrix = merged_df[numeric_cols].corr()
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm', fmt = '.2f')
plt.title('Correlation Heatmap of Key Metrics')
plt.show()
```

### Q19: Does taking training opportunities impact performance ratings?

```python
plt.figure(figsize = (12, 6))
sns.boxplot(data = merged_df, x = 'TrainingOpportunitiesTaken', y = 'ManagerRating', palette = 'vlag')
plt.title('Training Opportunities Taken vs Manager Rating')
plt.show()
```

### Q20: What is the relationship between years at company and years since last promotion?

```python
plt.figure(figsize = (12, 6))
sns.scatterplot(data = Emp, x = 'YearsAtCompany', y = 'YearsSinceLastPromotion', hue = 'Attrition', palette = 'husl', alpha =0.7)
plt.title('Years at Company vs. Years Since Last Promotion')
plt.show()
```

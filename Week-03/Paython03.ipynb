```python
# important Libraries
import pandas as pd # Preprocessing, I/O, Cleaning
import numpy as np # Linear Algebra (Matrices Manipulation)
import matplotlib.pyplot as plt # Visualization
import seaborn as sns
%matplotlib inline
from sklearn.ensemble import RandomForestClassifier
import warnings
warnings.filterwarnings('ignore')
```

```python
Emp = pd.read_csv('Employee.csv')
Edu = pd.read_csv('EducationLevel.csv')
Per = pd.read_csv('PerformanceRating.csv')
Rating = pd.read_csv('RatingLevel.csv')
Sat = pd.read_csv('SatisfiedLevel.csv')
```

```python
# Prepare Data for Predictive Modeling
latest_perf = Per.sort_values('ReviewDate').groupby('EmployeeID').tail(1)
df = pd.merge(Emp, latest_perf, on = 'EmployeeID', how = 'inner')
```

```python
# Convert Binary Columns to Numeric (1/0) for probabilities
df['Attrition_Num'] = (df['Attrition'] == 'Yes').astype(int)
df['OverTime_Num'] = (df['OverTime'] == 'Yes').astype(int)    
```

### Create Dashboard Figure (5 rows x 4 columns)

```python
fig, axes = plt.subplots(5, 4, figsize = (24, 25))
fig.suptitle('20 Predictive Analytics & Cross-Sectional Forecasting Questions', fontsize = 28, fontweight = 'bold', y = 0.92)
axes = axes.flatten()
sns.set_theme(style = "whitegrid")
```

### Q1: Attrition vs. Distance From Home (Logistic/Probability)

```python
sns.regplot(x = 'DistanceFromHome (KM)', y = 'Attrition_Num', data = df, logistic = True, ax = axes[0], scatter_kws = {'alpha':0.1}, line_kws = {'color':'red'})
axes[0].set_title('Predict Attrition by Commute Distance')
axes[0].set_ylabel('Probability of Attrition')
```

### Q20: Feature Importance for Attrition (Random Forest Classifier)

```python
features = ['Age', 'DistanceFromHome (KM)', 'Salary', 'YearsAtCompany', 'StockOptionLevel', 
            'JobSatisfaction', 'WorkLifeBalance', 'EnvironmentSatisfaction']
X = df[features].fillna(0)
y = df['Attrition_Num']
rf = RandomForestClassifier(random_state = 42, n_estimators = 100)
rf.fit(X, y)
importances = pd.Series(rf.feature_importances_, index = features).sort_values(ascending = False)
sns.barplot(x = importances.values, y = importances.index, ax = axes[19], palette = 'viridis', orient = 'h')
axes[19].set_title('Top Predictors of Attrition (Random Forest)')
```
